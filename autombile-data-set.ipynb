```python
# how many cars based on the 'body-style' column

df['body-style'].value_counts().to_frame()
```

```python
# how many cars are sedan, hatchback and wagon
spf_body_style_cars = ['sedan', 'hatchback', 'wagon']
new_table = df[ (df['body-style'].isin(spf_body_style_cars)) ]
new_table['body-style'].value_counts().to_frame()
```

```python
# how many cars are rear wheel vs front wheel and 4 wheel drives
# Based on and 'engine-location' 'drive-wheels'
spf_engin_location = ['rear', 'front']
new_table = df[ (df['engine-location'].isin(spf_engin_location)) & (df['drive-wheels']=='fwd') ]
new_table['engine-location'].value_counts().to_frame()
```
